### Imports

```python
# -------------------------------------------
# üåæ Crop Recommendation ‚Äî Data Preprocessing
# -------------------------------------------

# 1Ô∏è‚É£ Import libraries
import pandas as pd
import os
from sklearn.preprocessing import LabelEncoder

# 2Ô∏è‚É£ Load the raw dataset
raw_path = '../data/raw/raw_crop.csv'
data = pd.read_csv(raw_path)

print("‚úÖ Raw dataset loaded successfully!")
print("Shape:", data.shape)
data.head()
```

```python
# 3Ô∏è‚É£ Drop unnecessary columns
columns_to_drop = ['District_Name', 'Fertilizer', 'Link']
data.drop(columns=columns_to_drop, inplace=True)
print(f"Dropped columns: {columns_to_drop}")
```

```python
# 4Ô∏è‚É£ Encode categorical columns
le_soil = LabelEncoder()
le_crop = LabelEncoder()

data['Soil_color'] = le_soil.fit_transform(data['Soil_color'])
data['Crop'] = le_crop.fit_transform(data['Crop'])

print("‚úÖ Encoded categorical columns (Soil_color, Crop)")
data.head()
```

```python
# 5Ô∏è‚É£ Check for missing values
print("üß© Missing Values Summary:")
print(data.isnull().sum())
```

```python
# 6Ô∏è‚É£ Save processed dataset
processed_dir = '../data/processed'
os.makedirs(processed_dir, exist_ok=True)

processed_path = os.path.join(processed_dir, 'processed_crop.csv')
data.to_csv(processed_path, index=False)

print(f"üéâ Processed dataset saved successfully at: {processed_path}")
```
